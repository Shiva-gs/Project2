# New York Top Atractions

```python
# import library 
from bs4 import BeautifulSoup as bs
import requests
```

```python
# tripadviser url/retrieve page with the requests module/parse with 'html.parser
url = 'https://www.tripadvisor.com/Attractions-g60763-Activities-New_York_City_New_York.html'
response = requests.get(url)
soup = bs(response.text,'html.parser')
```

```python
# examine the data
print(soup.prettify())
```

```python
# name results are returned as an iterable list
result_name = soup.find_all('a', class_="attractions-attraction-overview-pois-PoiInfo__name--SJ0a4")
# print(result_name)

    
```

```python
# get all the name
#store the top 30 attraction into a list
title_list = []
for name in result_name:
    try:
        title = name.text
        
        if(title):
            
            title_list.append(title)
            print(title)
    except AttributeError as e:
        print(e)

print(f"\n\nTop {len(title_list)} attractions is listed here")
```

```python
# used the same methdo to obtain price for each attraction 
result_review = soup.find_all('span', class_="reviewCount styleguide-bubble-rating-BubbleRatingWithReviewCount__reviewCount--37tMc")
#print(result_review)
review_list = []
for review in result_review:
    review_list.append(review.text)

review_list_filtered = review_list[16:46]
print(review_list_filtered)
print(len(review_list_filtered))

```

```python
#store data to a object export to mongodb later
ny_top30 = {
    "attractions":title_list,
    "reviews":review_list_filtered
}
```
